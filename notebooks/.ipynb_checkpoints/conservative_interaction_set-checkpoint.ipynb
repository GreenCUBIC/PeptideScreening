```julia
BIOGRID_VERSION = "4.3.196"
BIOGRID_URL = "https://downloads.thebiogrid.org/Download/BioGRID/Release-Archive/BIOGRID-$BIOGRID_VERSION/BIOGRID-ALL-$BIOGRID_VERSION.tab3.zip"
```

```julia
# Unzip the data
if !isfile("../data/biogrid_$BIOGRID_VERSION.tab3.zip")
    run(`unzip ../data/biogrid_$BIOGRID_VERSION.tab3.zip -d $DESTINATION`)
end
```

```julia
# Load the interaction data
interactions_df = DataFrame(CSV.File("$DESTINATION/BIOGRID-ALL-$BIOGRID_VERSION.tab3.txt"))
```

```julia
# Print the column names
println("Columns are:")
println(names(interactions_df))
```

```julia
# Filter out non-human
filter!(row -> row["Organism ID Interactor A"] == 9606 && row["Organism ID Interactor B"] == 9606, interactions_df)
```

```julia
# Only keep the interactions discovered with stringent methods (high-quality interactions)
CONSERVATIVE_SYSTEMS = Set([
      "Two-hybrid",
      "Affinity Capture-MS",
      "Affinity Capture-Western",
      "Reconstituted Complex",
      "Affinity Capture-Luminescence",
      "Co-crystal Structure",
      "Far Western",
      "FRET",
      "Protein-peptide",
      "Co-localization",
      "Affinity Capture-RNA",
      "Co-purification"
])

filter!(row -> row["Experimental System"] âˆˆ CONSERVATIVE_SYSTEMS, interactions_df)
```

```julia
interactions = select(
    interactions_df[:, [
        "SWISS-PROT Accessions Interactor A", "SWISS-PROT Accessions Interactor B", "Publication Source"
    ]],
    "SWISS-PROT Accessions Interactor A" => "prot_a",
    "SWISS-PROT Accessions Interactor B" => "prot_b",
    "Publication Source" => "source"
)
```

```julia
# Count the number of times the interactions have been validated in different sources
sources = Dict()
pair_counts = Dict()

for pair in eachrow(interactions)
    p1 = min(pair["prot_a"], pair["prot_b"])
    p2 = max(pair["prot_a"], pair["prot_b"])
    source = pair["source"]
    
    pair_string = "$p1 $p2"
    
    if pair_string âˆ‰ keys(pair_counts)
        pair_counts[pair_string] = 1
        sources[pair_string] = [source]
    elseif pair_string âˆˆ keys(pair_counts) && source âˆ‰ sources[pair_string]
        pair_counts[pair_string] += 1
        push!(sources[pair_string], source)
    else
        continue
    end
end
```

```julia
pair_counts
```

```julia
# Only keep pairs that are present more than once (multiple lines of evidence)
duplicated_pairs = [split(k) for (k, v) in pair_counts if v >= 2]
```

```julia
# Save these interactions as conservative+ (these interactions will lead to overfitting,
# because some of the interactors are endogenous versions of the FDA-approved peptides.)
interactions_dir = "$DESTINATION/interactions"
if !isdir(interactions_dir)
    mkdir(interactions_dir)
end

open("$DESTINATION/interactions/homo_sapiens_with_endogenous_peptides.tsv", "w") do io
    write(io, join(["$(pair[1])\t$(pair[2])" for pair in duplicated_pairs], "\n"))
end
```
